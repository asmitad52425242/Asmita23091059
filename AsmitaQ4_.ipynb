```python

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(59)


coefficients = np.random.rand(6)


def polynomial(x):
    return np.polyval(coefficients[::-1], x)
x = np.linspace(-5, 5, 250)


y = polynomial(x)

plt.figure(figsize=(7, 5))
plt.plot(x, y, label='Polynomial')

# Find maxima and minima
maxima_x = x[np.argmax(y)]
maxima_y = np.max(y)
minima_x = x[np.argmin(y)]
minima_y = np.min(y)
plt.scatter([maxima_x, minima_x], [maxima_y, minima_y], color='red')
plt.text(maxima_x, maxima_y, f'Maximum\n({maxima_x:.2f}, {maxima_y:.2f})', ha='center', va='bottom')
plt.text(minima_x, minima_y, f'Minimum\n({minima_x:.2f}, {minima_y:.2f})', ha='center', va='top')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of Random 5th Degree Polynomial with Maxima and Minima')
plt.legend()
plt.grid(True)
plt.show()
```
